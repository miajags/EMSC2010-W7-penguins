### Help a lost penguin find its way home

* During a winter field trip to Antartica you find a lost penguin.
* It's dark and there is a snow storm!
* You cannot see enough to identify what type of penguin it is.
* But you can measure its height, which is 71 cm
* Should you point in the penguin in the direction of the Emperor or Adélie colony?

What you know:
Heights of the Emperor penguin population are normally distributed:
* Mean = 115 cm
* Standard deviation = 5 cm

Heights of the Adélie penguin population are normally distributed:
* Mean = 60 cm
* Standard deviation = 8 cm

You also have Bayes theorem to help!

$p(A|B) = \frac{p(B|A)⋅p(A)}{p(B)}$,

where:

* $p(A)$ is the prior probability that the penguin is species $A$.
* $p(B|A)$ is the likelihood that a species $A$ penguin has a height of 71 cm.
* $p(B)$ is the probability of observing a 71 cm tall penguin (we won't worry about this, I'll explain why later).
* $p(A|B)$ is the posterior probability that the penguin is species $A$ given that its height is $B$ (this is what we are interested in).


First we'll import the packages we need:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm #Scipy function for working with normal distributions
```

```python
prior_E = 0.5 #prior (likelohood) the penguin is an emperor penguin is 50/50
prior_A = 1 - prior_E #makes sure that the priors add up to 1

#prior can be changed e.g. prior_E=0.95 if we lived next to an emperor penguin colony. This will affect results.
```

```python
#plot normal distributions
height = np.linspace(40,140,1001) #equally spaced 1001 values between 40 and 140

#likelihood emperor penguin
likelihood_E = norm.pdf(height, loc=115, scale=5) #mean is 155cm and standard deviation of 5
plt.plot(height, likelihood_E)
plt.xlabel('Height (cm)')
plt.ylabel('PDF')

#likelihood adelie penguin
likelihood_A = norm.pdf(height, loc=60, scale=8)
plt.plot(height, likelihood_A)
plt.xlabel('Height (cm)')
plt.ylabel('PDF')

plt.title('Normal distribution of Emperor and Adelie penguins')
plt.legend(['Emperor', 'Adélie'])
plt.show()

#PDF = probability distribution function
```

```python
#what is my penguin? Likelihood.

height_penguin = 71
likelihood_E = norm.pdf(height_penguin, loc=115, scale=5)
likelihood_A = norm.pdf(height_penguin, loc=60, scale=8)
print(likelihood_E)
print (likelihood_A)
```

```python
#what is my penguin? Probability.

prob_E = likelihood_E * prior_E
prob_A = likelihood_A * prior_A


#in Bayes' theorem, we need the probability of 'B' being true - we need the probabilities to add up to 1

prob_total = prob_E + prob_A

prob_E = prob_E / prob_total
prob_A = prob_A / prob_total

print(prob_E)
print(prob_A)

#how do we intrepet this prosterier probability?
#prob_A is verrryyyy close to 1 - therefore my penguin, using Bayes' theorem to confirm, is probably an Adelie penguin
```
